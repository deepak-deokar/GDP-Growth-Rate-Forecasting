```python
# Step 1: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read evaluation results
results_df = pd.read_csv("../data/model_results.csv")

# Filter for India only (or any country you want to analyze)
india_results = results_df[results_df['Country'] == 'India']

print(india_results)
```

```python
# Visualization
plt.figure(figsize=(10,6))
plt.bar(india_results['Model'], india_results['RMSE'], color='skyblue')
plt.title('Model Comparison - RMSE (India)')
plt.ylabel('RMSE')
plt.grid(True)
plt.show()

plt.figure(figsize=(10,6))
plt.bar(india_results['Model'], india_results['MAE'], color='lightgreen')
plt.title('Model Comparison - MAE (India)')
plt.ylabel('MAE')
plt.grid(True)
plt.show()
```
